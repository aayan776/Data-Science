```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from ipywidgets import interact, IntSlider
from sklearn.datasets import make_classification
```

```python
iris = load_iris()
x = iris.data
y = iris.target
#Train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
#Scaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
#KNN
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
#Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification report:", classification_report(y_test, y_pred, target_names = iris.target_names))
```

```python
x, y = make_classification(
    n_samples = 200,
    n_features = 2,
    n_informative = 2,
    n_redundant = 0,
    n_repeated = 0,
    n_classes = 2,
    n_clusters_per_class = 1,
    random_state = 42
)
#Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2, random_state = 42)
#Plot
def plot(k):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(x_train, y_train)
  #Decision boundary
  x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
  y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
  #Meshgrid
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
  #Ravel
  Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)
  #Plot boundary
  plt.figure(figsize = (7,5))
  plt.contourf(xx, yy, Z, alpha = 0.3, cmap = 'winter')
  #Plot points
  plt.scatter(x_train[:, 0], x_train[:, 1], c = y_train, s = 70, cmap = plt.cm.winter, edgecolor = 'k', label = "Train")
  plt.scatter(x_test[:, 0], x_test[:, 1], c = y_test, s = 150, edgecolor = 'k', cmap = plt.cm.winter, marker = '*', label = "Test")
  plt.title(f"KNN Classification (k = {k})")
  plt.legend()
  plt.show()
#k slider
interact(plot, k = IntSlider(min = 1, max = 159, step = 1, value = 3))
```
