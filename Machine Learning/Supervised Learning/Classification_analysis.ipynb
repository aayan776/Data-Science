```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_breast_cancer
from google.colab import files

dataset = load_breast_cancer()
data = pd.DataFrame(dataset.data, columns = dataset.feature_names)
data['target'] = dataset.target
#Creating csv
csv_filename = "breast_cancer_dataset.csv"
data.to_csv(csv_filename, index = False)
print(f"Data save as {csv_filename}")
files.download(csv_filename)
```

```python
unique, counts = np.unique(data['target'], return_counts = True)
target_counts = pd.DataFrame({
    "Cancer_types" : dataset.target_names,
    "Count" : counts
})
import matplotlib as mpl
sns.set_style('darkgrid')
mpl.rcParams.update({"figure.figsize" : (8,6), "axes.titlepad" : 20.0, "axes.labelsize" : 14, "xtick.labelsize" : 12, "ytick.labelsize" : 12})
ax = sns.barplot(x = "Cancer_types", y = "Count", data = target_counts, hue = "Cancer_types", palette = "deep", legend = False)
# Titles and labels
ax.set_title('Distribution of Breast Cancer Target Classes', fontsize=18, weight='bold')
ax.set_xlabel('Cancer Type', fontsize=14)
ax.set_ylabel('Number of Samples', fontsize=14)

for p in ax.patches:
  height = p.get_height()
  ax.text(p.get_x() + p.get_width()/2., height + 5, f'{int(height)}', ha='center', fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
# Load dataset
dataset = load_breast_cancer()
X = dataset.data
y = dataset.target
feature_names = dataset.feature_names
target_names = dataset.target_names

x = dataset.data
y = dataset.target

feature_names = dataset.feature_names
target_name = dataset.target_names

x_feature = x[:, 0]
y_feature = x[:, 1]

plt.figure(figsize = (8,6))

colors = np.array(['red', 'green'])
labels = target_name

for target_value in np.unique(y):
  plt.scatter(
      x_feature[y == target_value],
      y_feature[y == target_value],
      c = colors[target_value],
      label = labels[target_value],
      edgecolor = 'k', alpha = 0.7, s = 60
  )

plt.title('Breast Cancer Dataset: Mean Radius vs Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
#Margin line
x_vals = np.linspace(min(x_feature), max(x_feature), 100)
y_vals = 15 - 0.1 * x_vals
plt.plot(x_vals, y_vals, color = "blue", linestyle = '--', linewidth = 2, label = "Margin Line")
plt.legend(title = "Cancer types")
plt.grid(True)
plt.show()
```
