```python
import nltk
from nltk.chat.util import Chat, reflections


reflections = {
  "i am"       : "you are",
  "i was"      : "you were",
  "i"          : "you",
  "i'm"        : "you are",
  "i'd"        : "you would",
  "i've"       : "you have",
  "i'll"       : "you will",
  "my"         : "your",
  "you are"    : "I am",
  "you were"   : "I was",
  "you've"     : "I have",
  "you'll"     : "I will",
  "your"       : "my",
  "yours"      : "mine",
  "you"        : "me",
  "me"         : "you"
}


pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how are you today?"]
    ],
    [
        r"hi|hey|hello",
        ["Hello!", "Hey there!"]
    ],
    [
        r"what are you|who are you|what is your name",
        ["I am a bot created by Tawseef Mamun to answer questions about the popular book series Percy Jackson"]
    ],
    [
        r"Who wrote percy jackson?",
        ["The author Rick Riordan wrote all the books set in the percy jackson universe."]
    ],
    [
        r"sorry (.*)",
        ["It's alright", "It's OK", "never mind"]
    ],
    [
        r"Who is rick riordan?",
        ["Rick Riordan is an American author who has written many fantasy novel series based on different mythologies"]
    ],
    [
        r"(.*) age?",
        ["As of the events of Trials of Apollo, Percy Jackson is 18 years of age."]
    ],
    [
        r"Where is camp Half-Blood located?",
        ["Camp Half-Blood is located on the north shore of Long Island, New York."]
    ],
    [
        r"Where is camp jupiter located?",
        ["Camp Jupiter, a Roman demigod training ground, is located in the Oakland Hills near San Francisco, California"]
    ],
    [
        r"(.*) created ?",
        ["Tawseef Mamun created me using Python's NLTK library", "Who knows?"]
    ],
    [
        r"How many books are there in the percy jackson series?",
        ["There are 5 books in the original series. 5 more for Heroes of Olympus and 5 more for Trials of Apollo making a total of 15 books."]
    ],
    [
        r"how is the weather in (.*)?",
        ["The weather in %1 is awesome like always", "Too hot here in %1", "Too cold here in %1", "Never even heard about %1"]
    ],
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy"]
    ],
    [
        r"Which team's logo does Annabeth's invisibility cap have?",
        ["Annabeth's invisibility cap has the logo of the Yankees baseball team on it"]
    ],
    [
        r"quit",
        ["Bye, take care.", "See you soon.", "It was nice talking to you. See you soon."]
    ],
]



def chat():
    print("Hi! I am a chatbot created by Codingal Edu. Pvt. Lim. for your service.")
    chatbot = Chat(pairs, reflections)
    chatbot.converse()

# Initiate the conversation
if __name__ == "__main__":
    chat()
```
