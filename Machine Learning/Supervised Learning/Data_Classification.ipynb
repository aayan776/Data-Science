```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_breast_cancer
from google.colab import files

data = pd.read_csv('insurance.csv')
data.head(10)
```

```python
data.shape
```

```python
x = pd.DataFrame(data['sex'])
x = x['sex'].apply(lambda x: 1 if x == 'male' else 0)
y = np.array(data[['bmi']])
color = ['red', 'green']
labels = ['male', 'female']

for target_values in np.unique(x):
  plt.scatter(
      x[x == target_values],
      y[x == target_values],
      c = color[target_values],
      label = labels[target_values],
      edgecolor = 'k', s = 60
  )
```
