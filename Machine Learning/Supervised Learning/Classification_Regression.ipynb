```python
df['released_year'] = pd.to_datetime(df['date_added'], errors = 'coerce').dt.year

plt.figure(figsize = (10,6))
sns.histplot(df['released_year'].dropna(), bins = 30, kde = False)
plt.title("Content added to Netflix per year")
plt.xlabel("Year")
plt.ylabel('Number of titles')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize = (10,8))
sns.barplot(x = top_countries.values, y = top_countries.index, palette = 'mako')
plt.title("Top 10 countries in terms of content count")
plt.xlabel("Number of titles")
plt.show()
```

```python
df_class = df[['type', 'duration', 'date_added']].dropna()

df_class['release_year'] = pd.to_datetime(df['date_added'], errors = 'coerce').dt.year

df_class = df_class.dropna()

df_class['type'] = df_class['type'].map({'Movie': 0, 'TV Show': 1})

df_class['duration'] = df_class['duration'].str.extract('(\d+)').astype(int)

df_class.head()
```

```python
#HW -------!!!!
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

x = df_class[['duration', 'release_year']]
y = df_class['type']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

clf = RandomForestClassifier()
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
print(classification_report(y_test, y_pred, target_names = ["Movie", "TV Show"]))
```
