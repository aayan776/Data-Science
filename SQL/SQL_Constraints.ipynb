```python
#Step 1 Link database
import sqlite3
database = "database.sqlite"
conn = sqlite3.connect(database)
cursor = conn.cursor()
print("Data opened successfully.")
```

```python
#Step 2 Create table
```

```python
cursor.execute("""CREATE TABLE IF NOT EXISTS Users (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  name TEXT NOT NULL,
  email TEXT UNIQUE,
  age INTEGER CHECK(age >= 18)
  )""")
print("Table created successfully")
conn.commit()
```

```python
#Step 3 Give valid info
try:
  cursor.execute("""INSERT INTO Users (name, email, age) VALUES (?, ?, ?) """,
   ("John Doe", "Johnemail@gmail.com", 28))
except sqlite3.IntegrityError as e:
  print("Error:", e)
print("Valid data inserted")
```

```python
#Step 4 Give valid info
try:
  cursor.execute("""INSERT INTO Users (name, email, age) VALUES (?, ?, ?) """,
   ("John Doe", "notJohnemail@gmail.com", 16))
except sqlite3.IntegrityError as e:
  print("Error:", e)
print("Invalid data inserted")
```
