```python
# Create a table
cursor = con.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT NOT NULL,
age INTEGER
);
""")
con.commit()
print("Table created successfully")
```

```python
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Alice", 10))
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Bob", 11))
con.commit()
print("Data inserted successfully")
```
