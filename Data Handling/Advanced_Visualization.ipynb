```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
weather = pd.read_csv('Weather Dataset - Trial Activity DataSet.csv.csv')
titanic = pd.read_csv("Titanic Dataset.csv")
```

```python
sns.barplot(x = weather['wind_speed'], y = weather['temperature'])
```

```python
#Distplot
sns.distplot(weather['temperature'])
```

```python
#Distplot - rug = true
sns.distplot(weather['humidity'], rug = True)
```

```python
#Jointplot
sns.jointplot(x = weather["humidity"], y = weather["temperature"])
```

```python
#Jointplot kind = hex
sns.jointplot(x = weather["humidity"], y = weather["temperature"], kind = "hex")
```

```python
#Jointplot - kde
sns.jointplot(x = weather["humidity"], y = weather["temperature"], kind = "kde")
```

```python
#Pairplot
sns.pairplot(weather[["humidity", "temperature", "air_pollution_index"]])
```

```python
#Strip plot
sns.stripplot(x = weather["weather_type"], y = weather["temperature"])
```

```python
#Strip plot - jitter = true
sns.stripplot(x = weather["weather_type"], y = weather["temperature"], jitter = True)
```

```python
#Swarm plot
plt.figure(figsize = (16, 16))
sns.swarmplot(x = weather["humidity"], y = weather["temperature"])
```

```python
#Box plot
plt.figure(figsize = (12, 12))
sns.boxplot(x = weather["humidity"], y = weather["temperature"],)
```

```python
#Box plot - hue
plt.figure(figsize = (12, 12))
sns.boxplot(x = weather["humidity"], y = weather["temperature"], hue = weather["weather_type"])
```

```python
#Count plot
sns.countplot(x = weather["weather_type"])
```

```python
#Count plot - hue
sns.countplot(x = weather["weather_type"], hue = weather["weather_type"])
```

```python
#Point plot - hue
plt.figure(figsize = (12, 12))
sns.pointplot(x = weather["humidity"], y = weather["temperature"], hue = weather["weather_type"])
```

```python
#Implot
plt.figure(figsize = (12, 12))
sns.lmplot(x = "humidity", y = "temperature", data = weather)
```

```python
#Implot - hue
plt.figure(figsize = (12, 12))
sns.lmplot(x = "humidity", y = "temperature", hue ="weather_type", data = weather)
```
