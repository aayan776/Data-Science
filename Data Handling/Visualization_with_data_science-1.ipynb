```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting aesthetics for clarity
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('country_vaccinations.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
#Heatmap - cbar False
subset = df.iloc[:5200:]
plt.figure(figsize = (12,8))
sns.heatmap(subset.isnull(), cbar = False, cmap = 'viridis')
plt.show()
```

```python
#Heatmap - cbar True
subset = df.iloc[:5200:]
plt.figure(figsize = (12,8))
sns.heatmap(subset.isnull(), cbar = True, cmap = 'viridis')
plt.show()
```

```python
#Drop every row with atleast 1 null value
df.dropna()
```

```python
#Drop every row with all null values
df.dropna(how = 'all')
```

```python
#Fill all null values with a designated value
df.fillna(0)
```

```python
#Fill all null values with a valid value
df.fillna(method = 'bfill')
```
