```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
df = pd.read_csv('Iris.csv')
data = pd.read_csv('Titanic Dataset.csv')
```

```python
#Quartiles and Interquartiles
q1pl = np.quantile(df['PetalLengthCm'], 0.25)
q2pl = np.quantile(df['PetalLengthCm'], 0.50)
q3pl = np.quantile(df['PetalLengthCm'], 0.75)
print("First quartile:", q1pl)
print("Second quartile:", q2pl)
print("Third quartile:", q3pl)
iqr_pl = q3pl - q1pl
print("InterQuartile Range:", iqr_pl)
```

```python
q1pw = np.quantile(df['PetalWidthCm'], 0.25)
q2pw = np.quantile(df['PetalWidthCm'], 0.50)
q3pw = np.quantile(df['PetalWidthCm'], 0.75)
print("First quartile:", q1pw)
print("Second quartile:", q2pw)
print("Third quartile:", q3pw)
iqr_pw = q3pw - q1pw
print("InterQuartile Range:", iqr_pw)
```

```python
q1sl = np.quantile(df['SepalLengthCm'], 0.25)
q2sl = np.quantile(df['SepalLengthCm'], 0.50)
q3sl = np.quantile(df['SepalLengthCm'], 0.75)
print("First quartile:", q1sl)
print("Second quartile:", q2sl)
print("Third quartile:", q3sl)
iqr_sl = q3sl - q1sl
print("InterQuartile Range:", iqr_sl)
```

```python
q1sw = np.quantile(df['SepalWidthCm'], 0.25)
q2sw = np.quantile(df['SepalWidthCm'], 0.50)
q3sw = np.quantile(df['SepalWidthCm'], 0.75)
print("First quartile:", q1sw)
print("Second quartile:", q2sw)
print("Third quartile:", q3sw)
iqr_sw = q3sw - q1sw
print("InterQuartile Range:", iqr_sw)
```

```python
q1age = np.quantile(data['Age'], 0.25)
q2age = np.quantile(data['Age'], 0.50)
q3age = np.quantile(data['Age'], 0.75)
print('Q1', q1age)
print('Q2', q2age)
print('Q3', q3age)
IQRage = q3age - q1age
print('IQR', IQRage)
```
