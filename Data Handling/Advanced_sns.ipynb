```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("Iris.csv")
data
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data["Species"])
```

```python
#Setting subplots
fig = plt.subplots(figsize = (12,8))
barWidth = 0.25
#Administering values
IT = [12,30,1,8,22,23,15,27,10,3]
ECE = [28,6,16,5,10,16,14,11,13,29]
CSE = [29,3,24,25,17,9,21,4,6,17]
#Set positions
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
#Plotting
plt.bar(br1, IT, color = 'r', width = barWidth, label = 'IT', edgecolor = 'grey')
plt.bar(br2, ECE, color = 'g', width = barWidth, label = 'ECE', edgecolor = 'grey')
plt.bar(br3, CSE, color = 'b', width = barWidth, label = 'CSE', edgecolor = 'grey')
#Xticks
plt.xlabel('Branch', fontweight = 'bold', fontsize = 16)
plt.ylabel('Students Passed', fontweight = 'bold', fontsize = 16)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015','2016','2017','2018','2019','2020','2021','2022','2023','2024'])
plt.legend()
plt.show()
```
