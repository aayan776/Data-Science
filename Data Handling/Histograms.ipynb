```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Dataset
data = np.array([20,12,76,89,43,
                 90,93,100,63,25,
                 33,47,59,18,44])
#Histogram
fig, ax = plt.subplots(figsize = (8,6))
ax.hist(data, bins = [0,20,40,60,80,100])
plt.show()
```

```python
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
#Creating dataset
np.random.seed(457698)
N_points = 10000
n_bins = 20
#Creating distribution
x = np.random.randn(N_points)
y = 0.8 ** x + np.random.randn(10000) + 25
#Creating histogram
figure, axis = plt.subplots(1, 1, figsize = (10,8), tight_layout = True)
axis.hist(x, bins = n_bins)
plt.show()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
type = [blood_sugar_men, blood_sugar_women]
colors = ['g', 'r']
labels = ['mens', 'women']
bins = [80, 100, 125, 150]
plt.xlabel('Blood sugar level')
plt.ylabel('Total no: of patients')

plt.hist(type, bins = bins, rwidth = 0.95, color=colors, label = labels, orientation = 'horizontal')
plt.title('Blood sugar level chart')
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
x = np.linspace(0,10,100)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,6))
x=np.linspace(0,10,100)
axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title('Sine')
axes[0,1].plot(x,np.cos(x))
axes[0,1].set_title('Cosine')
axes[1,0].plot(x,np.tan(x))
axes[1,0].set_title('Tangent')
axes[1,1].plot(x,np.log(x))
axes[1,1].set_title('Log')
plt.tight_layout()
plt.show()
```
