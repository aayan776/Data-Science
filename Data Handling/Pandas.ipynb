```python
import pandas as pd
import numpy as np
exam_data = {'name': ['Aayan', 'Anaya', 'Tawseef', 'Mamun', None, 'Samayra'],
             'score' : [88, 85, 76, 93, 71, None],
             'qualify' : [1, 1, None, 1, 1, 0]}
df=pd.DataFrame(exam_data)
print(df)
```

```python
#Drop rows with missing values
df.dropna(inplace = True)
print(df)
```

```python
#Reset index after dropping
df.reset_index(drop = True, inplace = True)
print(df)
```

```python
#Rename a column
df.rename(columns = {'score': 'marks'}, inplace = True)
print(df)
```
