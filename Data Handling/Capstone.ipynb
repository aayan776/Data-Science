```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

titan = pd.read_csv('Titanic Dataset.csv')
data = pd.read_csv('penguins_lter.csv')
df = pd.read_csv('penguins_size.csv')
```

```python
plt.figure(figsize = (10,8))
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
y = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
plt.plot(y, blood_sugar_men, color = 'blue', label = 'Men')
plt.plot(y, blood_sugar_women, color = 'red', label = 'Women')
plt.xlabel('Average blood sugar levels of both genders')
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
#Pie chart
data.groupby(titan['Survived']).size().plot(kind = 'pie', autopct = '%.2f')
plt.title('Survived - Died')
```

```python
data = {'C' : 20, 'C++' : 15, 'Java' : 25, 'JavaScript' : 31, 'Python' : 48, 'HTML' : 12, 'CSS' : 8, 'Data_Science' : 34}
courses = list(data.keys())
students = list(data.values())
fig = plt.figure(figsize = (10,8))
plt.bar(courses, students, color = 'black', width = 0.4)
plt.xlabel("Offered Courses")
plt.ylabel("Students Enrolled")
plt.title("Student enrollment in every course")
plt.show()
```

```python
#Setting subplots
fig = plt.subplots(figsize = (12,8))
barWidth = 0.25
#Administering values
IT = [12,30,1,8,22,23,15,27,10,3]
ECE = [28,6,16,5,10,16,14,11,13,29]
CSE = [29,3,24,25,17,9,21,4,6,17]
#Set positions
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
#Plotting
plt.bar(br1, IT, color = 'r', width = barWidth, label = 'IT', edgecolor = 'grey')
plt.bar(br2, ECE, color = 'g', width = barWidth, label = 'ECE', edgecolor = 'grey')
plt.bar(br3, CSE, color = 'b', width = barWidth, label = 'CSE', edgecolor = 'grey')
#Xticks
plt.xlabel('Branch', fontweight = 'bold', fontsize = 16)
plt.ylabel('Students Passed', fontweight = 'bold', fontsize = 16)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015','2016','2017','2018','2019','2020','2021','2022','2023','2024'])
plt.legend()
plt.show()
```
