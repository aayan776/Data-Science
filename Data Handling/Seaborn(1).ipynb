```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

house = pd.read_csv("/content/USA_Housing.csv")
#Round the values
house["Avg. Area Number of Bedrooms"] = house["Avg. Area Number of Bedrooms"].round()
print(house.columns)
#Bar graph
sns.barplot(x = "Avg. Area Number of Bedrooms", y = "Price", data = house)
plt.title("Average price of houses according to number of rooms")
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average price')
plt.show()
```

```python
sns.kdeplot(data = house["Price"], fill = True)
plt.title("Density of house prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```

```python
sns.kdeplot(data = house["Avg. Area Income"], label = "Income", fill = True)
sns.kdeplot(data = house["Price"], label = "Price", fill = True)
plt.title("Income vs House price distribution")
plt.xlabel("Money")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
sns.kdeplot(
            x = house["Avg. Area Income"],
            y = house["Price"],
            cmap = "Reds",
            fill = True,
            thresh = 0.05
            )
plt.title("2D KDE: Income vs Price")
plt.xlabel("Income")
plt.ylabel("Price")
plt.show()
```

```python
sns.displot(
    house["Price"], kde = True, height = 5, aspect = 1.5
)

plt.title("Distribution of house prices with KDE")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```
