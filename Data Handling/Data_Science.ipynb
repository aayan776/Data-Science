```python
a = np.arange(9, dtype = np.float64).reshape(3,3)
print("First array:")
print(a)

b = np.array([10,10,10])
print("Second array:")
print(b)

add = np.add(a,b)
print("Addition:")
print(add)

substract = np.subtract(a,b)
print("Substract:")
print(substract)

multiply = np.multiply(a,b)
print("Multiplication:")
print(multiply)

divide = np.divide(a,b)
print("Divide:")
print(divide)
```

```python
data_type = [("name", "S15"), ("class", "int"), ("height", "float")]
student_details = [("James", 5, 50.5), ("Brian", 6, 60.2), ("Hawk", 7, 75.7)]
arr = np.array(student_details, dtype = data_type)
print("Original array:")
print(student_details)
print("Sorted by height:")
print(np.sort(arr, order = "height"))
```
