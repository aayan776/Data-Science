```python
labels = ['Id', 'PetalLengthCm', 'PetalWidthCm', 'SepalLengthCm', 'SepalWidthCm']
for label in labels:
  print('Distribution of:', label)
  plt.boxplot(iris[label])
  plt.show()
```

```python
#Heatmap
sns.heatmap(df.corr(numeric_only = True), annot = True, cmap = 'mako')
```

```python
sns.heatmap(iris.corr(numeric_only = True), annot = True, cmap = 'terrain')
```

```python
#Distribution
labels = ['Id', 'PetalLengthCm', 'PetalWidthCm', 'SepalLengthCm', 'SepalWidthCm']
for label in labels:
  print('Distribution of:', label)
  sns.histplot(iris[label])
  plt.show()
```

```python
labels = ['Id', 'PetalLengthCm', 'PetalWidthCm', 'SepalLengthCm', 'SepalWidthCm']
for label in labels:
  print('Skewness of:', label)
  print('Value:', iris[label].skew(), '\n')
```
