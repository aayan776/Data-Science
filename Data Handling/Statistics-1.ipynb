```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statistics as stats
data = pd.read_csv('Iris.csv')
df = pd.read_csv('Titanic Dataset.csv')
```

```python
#Mean
sepal_length_mean = np.mean(data['SepalLengthCm'])
print("Sepal length in cm:", sepal_length_mean)
```

```python
sepal_width_mean = np.mean(data['SepalWidthCm'])
print("Sepal width in cm:", sepal_width_mean)
```

```python
petal_length_mean = np.mean(data['PetalLengthCm'])
print("Petal length in cm:", petal_length_mean)
```

```python
petal_width_mean = np.mean(data['PetalWidthCm'])
print("Petal width in cm:", petal_width_mean)
```

```python
#Median
sepal_length_median = np.median(data['SepalLengthCm'])
print("Sepal length median:", sepal_length_median)
```

```python
sepal_width_median = np.median(data['SepalWidthCm'])
print("Sepal width median:", sepal_width_median)
```

```python
petal_length_median = np.median(data['PetalLengthCm'])
print("Petal length median:", petal_length_median)
```

```python
petal_width_median = np.median(data['PetalWidthCm'])
print("Petal length median:", petal_width_median)
```

```python
sepal_length_mode = stats.mode(np.array(data['SepalLengthCm']))
print("Sepal length mode:", sepal_length_mode)
```

```python
sepal_width_mode = stats.mode(data['SepalWidthCm'])
print("Sepal width mode:", sepal_width_mode)
```

```python
petal_length_mode = stats.mode(data['PetalLengthCm'])
print("Petal length mode:", petal_length_mode)
```

```python
petal_width_mode = stats.mode(data['PetalWidthCm'])
print("Petal width mode:", petal_width_mode)
```
