```python
temp_mean = np.mean(df['Temperature (C)'])
print("Average temp in C:", temp_mean)
```

```python
for i in range(1, 13):
  month = df.loc[df['month'] == i]['Temperature (C)']
  print("For month:", i)
  print("Average temperature: " + str(np.mean(month)))
  print("Standard deviation of temperature: " + str(np.std(month)) + "\n")
```
