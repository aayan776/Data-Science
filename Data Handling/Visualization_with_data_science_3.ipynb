```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Countries Data.csv')
```

```python
data = {'C' : 20, 'C++' : 15, 'Java' : 25, 'JavaScript' : 31, 'Python' : 48, 'HTML' : 12, 'CSS' : 8, 'Data_Science' : 34}
courses = list(data.keys())
students = list(data.values())
fig = plt.figure(figsize = (10,8))
plt.bar(courses, students, color = 'black', width = 0.4)
plt.xlabel("Offered Courses")
plt.ylabel("Students Enrolled")
plt.title("Student enrollment in every course")
plt.show()
```

```python
#Setting subplots
fig = plt.subplots(figsize = (12,8))
barWidth = 0.25
#Administering values
IT = [12,30,1,8,22,23,15,27,10,3]
ECE = [28,6,16,5,10,16,14,11,13,29]
CSE = [29,3,24,25,17]
#Set positions
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
#Plotting
plt.bar(br1, IT, color = 'r', width = barWidth, label = 'IT', edgecolor = 'grey')
plt.bar(br2, ECE, color = 'g', width = barWidth, label = 'ECE', edgecolor = 'grey')
plt.bar(br3, CSE, color = 'b', width = barWidth, label = 'CSE', edgecolor = 'grey')
#Xticks
plt.xlabel('Branch', fontweight = 'bold', fontsize = 16)
plt.ylabel('Students Passed', fontweight = 'bold', fontsize = 16)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015','2016','2017','2018','2019'])
plt.legend()
plt.show()
```

```python
y1952 = df.loc[df['year'] == 1952]
y1952
```

```python
y2007 = df.loc[df['year'] == 2007]
y2007
```

```python
merge = y1952.merge(y2007, left_on = 'country', right_on = 'country')
merge
```

```python
merge.drop(['year_x', 'year_y'], axis = 1)
```

```python
merge['Population growth'] = merge['population_y'] - merge['population_x']
merge
```

```python
merge.shape, type(merge)
```

```python
merge = merge.sort_values('Population growth', ascending = False)
merge
```

```python
merge = merge.reset_index()
merge
```

```python
merge = merge.drop(['index'], axis = 1)
```

```python
merge = merge.head(10)
```

```python
names = ['China', 'India', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Bangledesh', 'Nigeria', 'Mexico', 'Philippines']
pop_grow = (merge['Population growth'] / 10 ** 6)
plt.figure(figsize = (12,8))
plt.bar(names, pop_grow, width = 0.6)
plt.xlabel('Country')
plt.ylabel('Population Growth (Millions)')
plt.title('Top 10 Countries w/the Biggest Population Growth from 1952 to 2007')
plt.xticks(rotation=45)

for x,y in zip(names, pop_grow):
  label = "{:.2f}".format(y)
  plt.annotate(label, (x,y), textcoords = 'offset points', xytext = (0,10), ha = 'center')
plt.show()
```
