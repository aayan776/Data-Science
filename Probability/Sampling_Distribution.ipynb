```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
puppies = np.array([1,1,0,1,0,0,1,0,1,0,1,1,1,0,0,1,0,0,1,0,0,1])

p = puppies.mean()
print('Mean:', p)
print('Standard Deviation:', puppies.std())
print('Variance:', puppies.var())
np.random.choice(puppies, size = (1,5), replace = True)
np.random.choice(puppies, size = (1,5), replace = True).mean()

sample_props = []
for i in range(10000):
  sample = np.random.choice(puppies, 5, replace = True)
  sample_props.append(sample.mean())
sample_props = np.array(sample_props)

sm = sample_props.mean()
print(sample_props)
print('Mean:', sm)
print('Standard Deviation:', sample_props.std())
print('Variance:', sample_props.var())

sample_props_20 = []
for i in range(10000):
  sample20 = np.random.choice(puppies, 20, replace = True)
  sample_props_20.append(sample20.mean())
sample_props_20 = np.array(sample_props_20)

sm_20 = sample_props_20.mean()
print(sample_props_20)
print('Mean:', sm_20)
print('Standard Deviation:', sample_props_20.std())
print('Variance:', sample_props_20.var())
```

```python
plt.hist(sample_props_20, bins = 20, color = 'skyblue')
plt.title('Sample mean distribution')
plt.xlabel('Mean of blue eyes')
plt.ylabel('Frequency')
plt.show()
```

```python
num = [1,10,50,100]
means = []
for j in num:
  np.random.seed(1)
  x = [np.mean(np.random.randint(-40, 40, j)) for _i in range(1000)]
  means.append(x)
k = 0

fig, ax = plt.subplots(2, 2,figsize = (6,6))
for i in range(0,2):
  for j in range(0,2):
    ax[i, j].hist(means[k], bins = 10, density = True)
    ax[i, j].set_title(label = num[k])
    k += 1
plt.show()
```
