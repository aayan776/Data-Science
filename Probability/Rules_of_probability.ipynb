```python
import random
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
set1 = {'A', 'B', 'G', 'F', 'K', 'U', 'O'}
set2 = {'A','F', 'J', 'D'}
union = set1.union(set2)
print(union)
total = list(union)
print(len(total))
```

```python
#Values
clicked_A = 380
clicked_B = 520
clicked_both = 150
total_customers = 1000
#Math
P_A = clicked_A/total_customers
P_B = clicked_B/total_customers
P_A_and_B = clicked_both/total_customers

P_A_or_B = P_A + P_B - P_A_and_B
P_A_times_B = P_A * P_B
#Professional Output
print(f'P(Clicked A)           = {P_A :.3f}')
print(f'P(Clicked B)           = {P_B :.3f}')
print(f'P(Both A and B)        = {P_A_and_B :.3f}')
print(f'P(Either A or B)       = {P_A_or_B :.3f}')
print(f'P(A) * P(B)            = {P_A_times_B :.3f}')
print(f"Independence           = {'Yes' if abs(P_A_times_B - P_A_and_B) < 0.01 else 'No'}")
print(f"Mutual Exclusivity     = {'Yes' if P_A_and_B == 0 else 'No'}")
#Venn
onlyA = clicked_A - clicked_both
onlyB = clicked_B - clicked_both
both = clicked_both

venn2(subsets = (onlyA, onlyB, both), set_labels = ('Clicked_A', 'Clicked B'))
plt.title('Venn Diagram: Email Campaign Clicks')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2
# Simulated dataset
np.random.seed(42)
data = pd.DataFrame({
'age': np.random.choice(['<50', '50-70', '>70'], size=1000, p=[0.3, 0.5, 0.2]),
'education': np.random.choice(['High School', 'Bachelor', 'Master'], size=1000, p=[0.4, 0.4, 0.2])})
print(data)
```

```python
#Values
High_school = len(data[data['education'] == 'High School'])
Over_70 = len(data[data['age'] == '>70'])
Pboth = len(data[(data['education'] == 'High School') & (data['age'] == '>70')])
total = len(data)
#Math
P_HS = High_school/total
P_70 = Over_70/total
P_both = Pboth/total

P_union = P_HS + P_70 - P_both
P_independent = P_HS + P_70
#Professional Output
print("=== Education vs Age of Death ===")
print(f"P(High School) = {P_HS:.3f}")
print(f"P(Age > 70) = {P_70 :.3f}")
print(f"P(Both HS & >70) = {P_both:.3f}")
print(f"P(Union) = {P_union:.3f}")
print(f"P(HS) * P(>70) = {P_independent:.3f}")
print(f"→ Independent? : {'Yes' if abs(P_independent - P_both) < 0.01 else 'No'}")
print(f"→ Mutually Exclusive? : {'Yes' if P_both == 0 else 'No'}")

onlyHS = High_school - Pboth
only70 = Over_70 - Pboth
venn2(subsets = (onlyHS, only70, Pboth), set_labels = ('High School', '70 +'))
plt.title('Venn Diagram: Education and Age')
plt.show()
```
