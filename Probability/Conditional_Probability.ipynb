```python
def a_and_b (a,b):
  if a == 1:
    prob_student = 0.3
    print(prob_student)
    if b == 1:
      prob_dining = 0.75
    else:
      prob_dining = 0.25
    print('Probability of b given a', prob_dining)
  if a == 2:
    prob_student = 0.7
    if b == 1:
      prob_dining = 0.6
    else:
      prob_dining = 0.4
    print('Probability of b given a', prob_dining)
  prob_a_and_b = prob_student * prob_dining
  return prob_a_and_b
print('Check the probability of any event occuring. First enter your choices:')
print('Is the student a freshman \n 1. Yes \n 2. No')
a = int(input('Enter choice: '))
print('Is the student a eating in the dining hall \n 1. Yes \n 2. No')
b = int(input('Enter choice: '))
result = a_and_b(a,b)
print('Here are the probabilities:', result)
```

```python
def prob_a_and_b(a, b, total):
  #Probability of singular outcome
  prob_a = a/total
  prob_bga = b/(total - 1)
  #Probability of both events ocurring
  prob_both = prob_a * prob_bga
  return prob_both
#Inputs
a = int(input('Enter number of red balls: '))
b = int(input('Enter number of blue balls: '))
total = a + b
print('Probability of pulling a and then b:', prob_a_and_b(a,b,total))
```
