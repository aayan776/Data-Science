```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2
```

```python
# Simulated dataset
np.random.seed(42)
data = pd.DataFrame({
'age': np.random.choice(['<50', '50-70', '>70'], size=1000, p=[0.3, 0.5, 0.2]),
'education': np.random.choice(['High School', 'Bachelor', 'Master'], size=1000, p=[0.4, 0.4, 0.2])})
print(data)
```

```python
#Values
High_school = len(data[data['education'] == 'High School'])
Over_70 = len(data[data['age'] == '>70'])
Pboth = len(data[(data['education'] == 'High School') & (data['age'] == '>70')])
total = len(data)
#Math
P_HS = High_school/total
P_70 = Over_70/total
P_both = Pboth/total

P_union = P_HS + P_70 - P_both
P_independent = P_HS + P_70
```

```python
crosstab = pd.crosstab(data['education'], data['age'], normalize = 'index')
plt.figure(figsize = (8,6))
sns.heatmap(crosstab, annot = True, cmap = 'YlGnBu')
plt.xlabel('Age Group')
plt.ylabel('Education')
plt.title('Probability Distribution: Age Group by Education')
plt.show()
```
