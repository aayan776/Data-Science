```python
prob_6 = stats.poisson.pmf(6,10)
print('Probability of it raining on six different days:', prob_6)

prob_12_14 = stats.poisson.pmf(12,10) + stats.poisson.pmf(13,10) + stats.poisson.pmf(14,10)
print('Probability of it raining on 12 - 14 different days:', prob_12_14)
```

```python
prob_20_call = 1 - stats.poisson.cdf(20, 15)
print('Probability of recieving 20 calls:', prob_20_call)

prob_21_17_call = stats.poisson.cdf(21, 15) - stats.poisson.cdf(16,15)
print('Probability of recieving 21 - 17 calls:', prob_21_17_call)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 5
p = 0.7
x = np.arange(0, n + 1)

pmf_values = stats.binom.pmf(x,n,p)
cdf_values = stats.binom.cdf(x,n,p)

prob_ex_3 = stats.binom.pmf(3,n,p)
print('Probability of it raining exactly 3 days', round(prob_ex_3, 4))

prob_to_3 = stats.binom.cdf(3,n,p)
print('Probability of it raining atmost 3 days', round(prob_to_3, 4))

prob_le_4 = 1 - stats.binom.cdf(4,n,p)
print('Probability of it raining atleast 4 days', round(prob_le_4, 4))
```

```python
plt.figure(figsize = (10,5))
plt.subplots = (1,2,1)
plt.bar(x, pmf_values, color = 'skyblue', edgecolor = 'black')
plt.xlabel('PMF: Number of rainy days')
plt.ylabel('Probability (x = k)')
plt.title('PMF: Exactly k rainy days')
plt.xticks(x)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
```
